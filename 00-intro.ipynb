## Informacin estructurada y no estructurada

Esencialmente existen dos tipos de informacin, **la estructurada** y la **no estructurada**. Los datos estructurados tienen un formato fijo que facilita su lectura y su procesamiento. Por ejemplo, las bases de datos relacionales de un sistema de informacin

Por otro lado, a la informacin que no posee elementos como los de las bases de datos se le conoce como informacin no estructurada y puede tener formatos diferentes. Por ejemplo los documentos escritos, ya que en su interior pueden contener cualquier tipo de informacin. 



## Anlisis de datos 

- El anlisis de los datos puede ser simple como calcular histogramas o ms elaborados como tcnicas de agrupamiento o clasificacin de datos, encontrar correlaciones, etc.
- Cada objeto estudiado se describe mediante **variables** o atributos que representan sus propiedades.
- Todas las variables de un objeto se llama **registro**.
- A la coleccin de todos los registros se le conoce como **conjunto de datos**, que generalmente se representa por una tabla donde cada fila es un registro y las columnas son los atributos.

Existen diferentes tipos de variables, segn el tipo de datos que contienen. Los tipos ms comunes son:
- Variable nominal. Indica una categora a la que pertenece el objeto. La categora pueden ser palabras o nmeros. Sus elementos no tienen un orden. 
- Variable binaria. Toma los valores de 0  1.
- Variables Ordinales. Los valores que toma tienen un orden.
- Variables de intervalo. Son variables numricas donde cada nmero representa un intervalo.


## Aprendizaje Supervisado

- En el aprendizaje supervisado se tiene los datos y cada uno de ellos tiene asociado un valor objetivo. 

- La **clasificacin** es uno de los problemas ms comunes en el aprendizaje supervisado. En l, se tiene que asignar la etiqueta correcta a los datos. 

- Los algoritmos de aprendizaje se alimentan de una gran coleccin de datos que ya fueron previamente etiquetados por expertos. Los algoritmos buscan patrones que les permitan clasificar de manera correcta datos nunca antes vistos ellos. A este proceso se le llama **entrenamiento**.

## Aprendizaje no supervisado

- En el aprendizaje no supervisado, se tienen los datos como en el caso anterior, pero no tienen asociado ningn valor.
- El objetivo entonces es agrupar los datos parecidos y diferenciarlos del resto. A este proceso se le llama **clustering**. 

- Los clusters formarn una particin de los datos y cada cluster es un conjunto que contiene objetos similares entre ellos pero distintos a los que hay en los dems clusters. 

- Por ejemplo, tenemos un conjunto de datos que contiene artculos de noticias provenientes de diarios nacionales. Slo tenemos los textos pero no estn etiquetados o separados por tema. El objetivo es crear un algoritmo que sea capaz de separarlos por tema, que ponga todos los de poltica en un conjunto, en otro los de deportes, y as para los temas de finanzas, internacionales y salud.
